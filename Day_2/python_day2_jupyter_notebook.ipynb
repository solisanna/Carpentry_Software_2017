```python
def celsius_fahrenheit(C):
    # F = (C * 9/5) + 32
    fahrenheit = (C * (9/5)) + 32
    return fahrenheit
```

```python
celsius_fahrenheit(100)
```

```python
celsius_fahrenheit(C=100)
```

```python
def fahrenheit_celsius(F):
    # C = (F - 32) * (5/9)
    celsius = (F - 32) * (5/9)
    return celsius
```

```python
fahrenheit_celsius(212)
```

```python
fahrenheit_celsius(71.6)
```

```python
# F=10 is the default if we were to not give it an input
def fahrenheit_celsius(F=10):
    # C = (F - 32) * (5/9)
    celsius = (F - 32) * (5/9)
    return celsius
```

```python
fahrenheit_celsius()
```

```python
fahrenheit_celsius(10)
```

```python
fahr = fahrenheit_celsius(F = 35)
```

```python
def split_county_by_pop(df, threshold):
    urban = df.loc[df['Population_Density'] >= threshold, :]
    rural = df.loc[df['Population_Density'] < threshold, :]
    return [urban, rural]
```

```python
urban_rural_counties = \
split_county_by_pop(dem_health_data, 500)
```
